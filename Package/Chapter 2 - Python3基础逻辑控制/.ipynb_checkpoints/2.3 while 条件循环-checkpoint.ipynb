## 语法: while 条件： ……[else ……]

```python
a = 32
```

```python
while a:  #死循环
    print(a)
```

```python
while a:  
    print(a)
    a = a -1
```

```python
a
```

```python
a = 32
while a:  
    print(a)
    a = a -1
else:
    print('end:', a)
```

## while和for循环有什么区别
#### while循环
重点是条件循环，一开始进入的时候就要对条件进行判断

#### for循环
循环条件是明确的集合，循环次数是明确的

### 关键字：break， 结束当前循环
### 关键字：continue， 跳过本次循环

```python
b = "dfakljfklsadklz"
for i in b:
    if i == 'k':
        print('跳过本次循环')
        continue
        print(i, '跳过')
    else:
        print("输出：",i)
```

```python
op = [123,345,567,67876,32432,34,45,88756,3754,9756,98,245]
```

```python
len(op)
```

```python
while op:
    p = op.pop()
    print("当前值：", p)
    if p > 50000:
        print("当前值大于50000，结束当前循环")
        break
```
